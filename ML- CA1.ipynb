```python
#fill in numeric column with means
df[['MasVnrArea','LotFrontage','GarageYrBlt']] = df[['MasVnrArea','LotFrontage','GarageYrBlt']].fillna(df.mean())
df
```

```python
#fill categorical variables with mode if missing less than 50%
df[['GarageType','BsmtFinType2','BsmtFinType1','GarageQual','GarageCond','FireplaceQu','GarageFinish']] = df[['GarageType','BsmtFinType2','BsmtFinType1','GarageQual','GarageCond','FireplaceQu','GarageFinish']].fillna(df.mode())
```

```python
#handling missing values by dropping columns with more than 50% of missing
df.drop(['MiscFeature', 'Fence', 'PoolQC','Alley'], axis=1, inplace=True)
```
