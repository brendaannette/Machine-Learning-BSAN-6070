```python
#drop labels and movie ID because they are not in the list for "The Post" movie in assignment
moviedata.head()
moviedata.drop(['Movie ID', 'Label'], axis=1, inplace=True)          
```

```python
# split the data into two parts: movie titles and remaining columns because we cannot use a categorical feature to scale the data and use KNN for the reccomender system
moviename = moviedata["Movie Name"]
moviedata1 = moviedata.drop("Movie Name", axis=1)
```

```python
#now scale the data for the numerical columns for KNN reccomender
# scale the remaining columns using StandardScaler
from sklearn.preprocessing import StandardScaler

#create an instance of StandardScaler
scaler = StandardScaler()

scaled_data = scaler.fit_transform(moviedata1)
```

```python
#implement the KNN algorithm
```

```python
#import KNN
import sklearn.neighbors
```

```python
#import nearestneighbors
from sklearn.neighbors import NearestNeighbors
```

```python
moviedata1.head()
```

```python
#use KNeighbors and NearestNeighbors
#use nearestneighbors = 6 because it is including 'The Post' movie as a neighbor
knn = NearestNeighbors(n_neighbors=6)
#using the entire properly encoded data sit to use the knn model on
knn.fit(scaled_data)
#this query_movie information is given from the CA05 assignment so we know what the nearest neighbors need to be similar too to reccomend similar movies
query_movie = [[7.2, 1, 1, 0, 0, 0, 0, 1]]
# istances of the 6 nearest neighbors of 'The Post' movie, and indices contains the indices of these neighbors in the dataset. 
distances, indices = knn.kneighbors(query_movie)
#indices are the actual indices of the 6 nearest neighbors of the 30 in the array nearest to query_movie (data representation of The Post)
```

```python
# display the recommendations
print("Top 5 Recommendations for 'The Post':")
#loop the indices array, the indices of nearest neighbors, getting the length of that to get the NN
for i in range(1, len(indices[0])):
    index = indices[0][i]
#the indices correspond to indices in the movie name array 
    movie_title = moviename.iloc[index]
    print(f"{i}. {movie_title}")
```
